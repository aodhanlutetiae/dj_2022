## Jomec staff

```python
url = 'https://www.cardiff.ac.uk/journalism-media-and-culture/people/academic-staff'

req = requests.get(url)
req.status_code
```

```python
soup = bs(req.content)
```

```python
soup
```

```python

for x in soup.find_all(class_ = 'profile-subtitle'):
    print(x.text)
    
```

```python
# print all the elements and nothing extra


for x in soup.find_all('div', class_ = 'profile profile-student with-image vcard'):
    for a in x.find_all('h2'):
        print(a.text)
    for b in x.find_all('p'):
        print(b.text)
    for c in x.find_all('dl'):
        for e in c.find('a'):
            print(e.text)
    for f in x.find_all('dd', class_ = 'profile-contact-telephone'):
        print(f.text)

    print('---')
```

```python
# assign the data to variables and use an empty list on each loop to store each loop's data

big_list_list = []

for x in soup.find_all('div', class_ = 'profile profile-student with-image vcard'):
    this_person = []
    for a in x.find_all('h2'):
        name = a.text
        this_person.append(name)
    for b in x.find_all('p'):
        job = b.text
        this_person.append(job)
    for c in x.find_all('dl'):
        for e in c.find('a'):
            email = e.text
            this_person.append(email)
    for f in x.find_all('dd', class_ = 'profile-contact-telephone'):
        tel = f.text
        this_person.append(tel)
        
    # append this loop's list to the big list
    big_list_list.append(this_person)
```

```python
# what does the big list look like?

big_list_list
```

```python
# same idea, but building a list of dictionaries

big_list_dict = []

for x in soup.find_all('div', class_ = 'profile profile-student with-image vcard'):
    person_dict = {}
    for a in x.find_all('h2'):
        name = a.text
        person_dict['name'] = name

    for b in x.find_all('p'):
        job = b.text
        person_dict['job'] = job

    for c in x.find_all('dl'):
        for e in c.find('a'):
            email = e.text
            person_dict['email'] = email

    for f in x.find_all('dd', class_ = 'profile-contact-telephone'):
        tel = f.text
        person_dict['tel'] = tel

    big_list_dict.append(person_dict)
    
```

```python
# what does the list of dictionaries look like?

big_list_dict
```

```python
# build a dataframe from the list of lists

cols = ['person', 'position', 'email', 'phone']

pd.DataFrame(big_list_list, columns = cols)
```

```python
# build a dataframe from the list of dicts

pd.DataFrame(big_list_dict)
```

## Council - single scraper

> only runs first five letters and only runs first page for each letter

```python
my_headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:71.0) Gecko/20100101 Firefox/71.0; Aidan ODonnell/odonnella4@cardiff.ac.uk. Teaching webscraping for journalism"}

# make a list of pages

letters_list = ['A', 'B', 'C', 'D', 'E']
pages_list = []
nb=0

for x in range(len(letters_list)):
    letter = letters_list[nb]
    url = f'https://www.coventry.gov.uk/directory/45/a-to-z/{letter}'
    pages_list.append(url)
    nb = nb + 1

# make a list of links on each page

link_list = []

for x in pages_list:
    
    base = 'https://www.coventry.gov.uk'
    
    url = x    
    req = requests.get(url, my_headers, timeout = 5)
    sleep(1)
    
    soup = bs(req.content)
    for y in soup.find_all('ul', class_ = 'list list--record'):
        for z in y.find_all('a'):
            end = z.attrs['href']
            link = base + end
            link_list.append(link)
    
data_list = []

for x in link_list:
    
    page_info = []

    url = x  
    req = requests.get(url, my_headers, timeout = 5)
    sleep(2)
    soup = bs(req.content)

    # collect the name of the FOI request
    for x in soup.find_all('h1'):
        title = x.text.strip()
        page_info.append(title)

    for x in soup.find_all('dd', class_ = 'definition__content definition__content--text-box'):
        # add .strip() to remove white space etc
        dates = x.text.strip()
        page_info.append(dates)

    for x in soup.find_all('dd', class_ = 'definition__content definition__content--select-box'):
        outcome = x.text.strip()
        page_info.append(outcome)

    for x in soup.find_all('dd', class_ = 'definition__content definition__content--link'):
        link = x.text.strip()
        page_info.append(link)
    
    data_list.append(page_info)
    
# make df from list of pages visited
cols = 'topic reference request_date response_date closed_date category outcome link'.split()

df = pd.DataFrame(data_list, columns = cols)

df
```
