```python
import requests
from bs4 import BeautifulSoup as bs
from time import sleep
```

```python
# can we print the letter pages with an f-string that we could loop?

nb = 1
pg = f'https://www.coventry.gov.uk/directory/45/a-to-z/A?page={nb}'
print(pg)

req = requests.get(pg)
print(req.status_code)
```

```python
# what gets returned when we get to a page number that doesn't exist?

nb = 200
pg = f'https://www.coventry.gov.uk/directory/45/a-to-z/A?page={nb}'
print(pg)
req = requests.get(pg)

print(req.status_code)
```

```python
# make a list to collect all the letter pages
letters_page_list = []

nb = 1
# assume that no letter has more than 100 pages
for x in range(100):
    
    # assemble the url & print it 
    pg = f'https://www.coventry.gov.uk/directory/45/a-to-z/A?page={nb}'
    print(pg)
    # request the page content
    req = requests.get(pg)
    print(req.status_code)
    
    # on each loop we check that we have got a 200 response to our request, if not we stop the loop
    if req.status_code != 200:
        break
    
    # otherwise add the page to the list
    else:
        letters_page_list.append(pg)
    
    nb = nb+1
    sleep(2)
    
print('finished')
print('nb of pages for letter A:', len(letters_page_list))    
```

```python

my_headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:71.0) Gecko/20100101 Firefox/71.0; Aidan ODonnell/odonnella4@cardiff.ac.uk. Teaching webscraping for journalism"}
letters_page_list = []

nb = 1
for x in range(100):
    
    pg = f'https://www.coventry.gov.uk/directory/45/a-to-z/A?page={nb}'
    
    req = requests.get(pg)
    
    if req.status_code != 200:
        break
    else:
        letters_page_list.append(pg)
    
    nb = nb+1
    sleep(2)
    
print('finished finding all the letter pages') 

link_list = []

# define the start of the links so we can construct full links
base = 'https://www.coventry.gov.uk'

# use the list of pages collected and request each one
for x in letters_page_list:
    
    url = x    
    req = requests.get(url, my_headers, timeout = 5)
    sleep(1)
    
    # from the soup collect all the links on the page
    soup = bs(req.content)
    for y in soup.find_all('ul', class_ = 'list list--record'):
        for z in y.find_all('a'):
            end = z.attrs['href']
            link = base + end
            link_list.append(link)

print(len(link_list), 'foi requests for this letter')
```
